```python
import pandas as pd
import numpy as np
df=pd.read_csv('healthcare_dataset.csv')
df
```

```python
df.info()
```

```python
#  DischargeDate and  DateofAdmission  in object datatype as they are in string format hence here we need to handle them....!
# converting it in to numerical form
```

```python
df['Year of Admission']=pd.to_datetime(df['Date of Admission']).dt.year
df['Month of Admission']=pd.to_datetime(df['Date of Admission']).dt.month
df['Day of Admission']=pd.to_datetime(df['Date of Admission']).dt.day     #Creating new three columns for Date of Admission
```

```python
df.drop(columns=['Date of Admission'],inplace=True)   # deleting the existing coloum
```

```python
df['Year of Discharge']=pd.to_datetime(df['Discharge Date']).dt.year
df['Month of Discharge']=pd.to_datetime(df['Discharge Date']).dt.month
df['Day of Discharge']=pd.to_datetime(df['Discharge Date']).dt.day     #Creating new three columns for that Discharge Date
```

```python
df.drop(columns=['Discharge Date'],inplace=True)  # deleting the existing coloum
```
