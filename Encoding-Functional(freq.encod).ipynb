```python
#In frequency encoding, categories are replaced with the frequency of their occurrences in the dataset.
#This means that a category appearing more frequently gets a higher value.
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('cars.csv')
df
```

```python
owner=df['owner'].value_counts()    #Storing value count of owner coloum in variable owner
df['owner_new']=df['owner'].map(owner)  #Mapping that value count of owner coloum By creating a new column called owner_new
```

```python
df['owner_new'] #displaying the coloum

# ie=0 5289----First owner
#    1 2105----Second owner
#    2 555-----third owner
# 8124 174-----Fourth & Above Owner    #These all are the value count and they are mapped 

#this is nothing but a frequency encoding
```

```python
df.groupby('owner')['selling_price'].mean()  #Grouping with selling price and their mean
```

```python
df['brand'].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder  #nominal data ke liye
OHE=OneHotEncoder(sparse_output=False,dtype='int32',handle_unknown='ignore') #object 

# sparse_output=False: The output will be a regular array, not a compressed one (sparse_matrix 0&1 matrix).
# dtype='int32': The data will be stored as 32-bit integers (a smaller, efficient number type).
# handle_unknown='ignore': If there are unknown values during encoding, they will be ignored, not causing errors.
```

```python
OHE.fit_transform(df[['brand']])
```

```python
pd.DataFrame(OHE.fit_transform(df[['brand']]))   #32 COLOUMS high dimension hence we can use frequency encoding
```

```python
df['brand']=df['brand'].map(df['brand'].value_counts()) #This is called frequency Encoding
#Mapping the column with their actual value counts that how much time they are occurring(frequency)
#And changing the column in that way is called frequency encoding 
```
