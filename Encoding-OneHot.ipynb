```python
import pandas as pd
import numpy as np

df = pd.read_csv('cars.csv')
df.tail(10)
```

```python
from sklearn.preprocessing import OneHotEncoder  #nominal data ke liye
```

```python
OHE=OneHotEncoder(sparse_output=False,dtype='int32',handle_unknown='ignore') #object 

# sparse_output=False: The output will be a regular array, not a compressed one (sparse_matrix 0&1 matrix).
# dtype='int32': The data will be stored as 32-bit integers (a smaller, efficient number type).
# handle_unknown='ignore': If there are unknown values during encoding, they will be ignored, not causing errors.
```

```python
own=OHE.fit_transform(df[['owner']])
```

```python
own #owner coloum encoding---> 4 values hence 4 columns in owner coloum
#Not directly like Ordinal encoder 0 1 2 3 Because the data is not in order form 
# Hence nominal data we use One hot encoder
```

```python
pd.DataFrame(own,columns=['First Owner', 'Second Owner', 'Third Owner',
                          'Fourth & Above Owner', 'Test Drive Car'])
# Creates a DataFrame (a table of data) using the own data, with the specified column names.
```

```python
x_train=[['satosa'],['verginica'],['versicolor']]
x_test=[['satosa'],['verginica'],['flower']]            #data frame data set
#x_test=[['satosa'],['verginica'],['versicolor']]  
# If we change xtest the mathematical encoding also changes as seen below                                              
```

```python
OHE.fit(x_train)  #training
```

```python
x_train=OHE.transform(x_train)
x_test=OHE.transform(x_test)        # transforming---->one hot encoding(nominal data)
```
