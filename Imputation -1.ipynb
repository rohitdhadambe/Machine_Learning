```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('train_data.csv')
```

```python
df
```

```python
df.drop(columns=['Name','Ticket','Cabin','Parch','PassengerId'],inplace=True) #drop meance delete the coloumn
#The inplace=True argument means that the changes are applied directly to df without creating a new DataFrame.
```

```python
df.info()
```

```python
# object means categorical data
```

```python
# imputation on coloumn Embarked and Age---------------->
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()
df['Age']=si.fit_transform(df[['Age']])          # imputation on coloumn age
```

```python
sim=SimpleImputer(strategy='most_frequent')
df['Embarked'].values.reshape(-1,1)              # imputation on coloumn Embarked
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df[['Sex']])
```

```python
oe=OrdinalEncoder(categories=[['S','C','Q']])
df['Embarked']=oe.fit_transform(sim.fit_transform(df[['Embarked']]))
```
