```python
# standardisation and normalisation
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = [
    col for col in df.columns 
    if df[col].dtype in ['float64', 'int64'] and col not in ['id', 'Profession', 'Degree', 
                                                             'Have you ever had suicidal thoughts ?',
                                                             'Family History of Mental Illness', 'Depression']
]

```

```python
scaler = StandardScaler()


# Fit and transform the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df
```

# Feature Selection

```python
numerical_colss = df.columns.tolist()  # bunch the all coloums
```

```python
from sklearn.feature_selection import VarianceThreshold
fe=VarianceThreshold(threshold=0.1)

selected_features = fe.fit_transform(df[numerical_colss])
mask = fe.get_support()


# Retrieve the names of selected features
selected_feature_names = [col for col, is_selected in zip(numerical_colss, mask) if is_selected]


# Create a new dataframe with the selected features
df_selected = pd.DataFrame(selected_features, columns=selected_feature_names, index=df.index)
```

```python
variances = pd.Series(fe.fit(df[numerical_colss]).variances_, index=numerical_colss)
print(variances)
```

```python
print("Original Data Shape:", df[numerical_colss].shape)
print("Selected Data Shape:", df_selected.shape)
```

```python
# Get the original column names
original_columns = df.columns.tolist()

# Get the names of the selected columns
selected_columns = [col for col, is_selected in zip(original_columns, fe.get_support()) if is_selected]

# Find the removed columns by comparing the original and selected columns
removed_columns = list(set(original_columns) - set(selected_columns))

print("Removed Columns:", removed_columns)
```

 Model selection == logistic regression


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression



X = df.drop(columns=['Depression'])  # Features (all columns except target)
y = df['Depression']               # Target variable


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=420)


model = LogisticRegression()
model.fit(X_train, y_train)

# When you use solver='liblinear',
# you're telling the model to use this fast, efficient method to find the best parameters for binary classification problems.

# class_weight='balanced' is useful when you have imbalanced classes, as it adjusts the model 
# to pay more attention to the minority class.

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
