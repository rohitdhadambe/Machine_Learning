```python
from sklearn.preprocessing import MinMaxScaler
MINMAX = MinMaxScaler()
MINMAX.fit(df[['Age']]) 
MINMAX.transform(df[['Age']])    #this is called scaling

# This is Normalization (range = 0 to 1) (depends on min and max val)  
```
