```python
import pandas as pd
import numpy as np

df = pd.read_csv('train_data.csv')
```

```python
df.drop(columns=['PassengerId','Name','Cabin','Ticket'],inplace=True)
```

```python
df.info()  # 714 missing values
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
# Mean Imputation: replace missing values in column 'A' with the mean of 'A'
```

```python
df.info()  # 0 missing values
```

```python
#Same we can do the median imputation

## Median Imputation: replace missing values in column 'B' with the median of 'B'
#df['B'].fillna(df['B'].median(), inplace=True)
```

```python
from sklearn.impute import SimpleImputer

# Initialize the SimpleImputer with mean strategy
si = SimpleImputer(strategy='mean')

# Apply mean imputation on the 'Age' column
mean_age1 = si.fit_transform(df[['Age']])
mean_age1


#u can directly do with the help of SimpleImputer from sklearn.impute
#same for median
```
